# **3. Compute and plot Discrete Fourier Transform (DFT) of a signal using DFT equation.**

**Import Libraries**

```python
import matplotlib.pyplot as plt
import numpy as np
import cmath
```

**DFT**

```python

Fs = 225   #Sample Frequency
t= np.arange(0,1,1/Fs)
f1=40     #frequencies of a complex signal                                        
f2=100
f3=80
x=np.sin(2*np.pi*f1*t)+np.cos(2*np.pi*f2*t)+np.sin(2*np.pi*f3*t) #complex signal with many frequencies


plt.subplot(3,1,1) #plotting of input signal
plt.title('Fig 1: Input Signal of time domain x(n)')
plt.plot(t,x)
plt.ylabel('Amplitude')
plt.xlabel('Time')


N=np.size(t)   #size of the discrete signal

#DFT
#Let X is the output of DFT
X=np.zeros(N,complex)
n=np.arange(N)
for k in range(N):
  X[k] = np.sum(x*np.exp(-2j*np.pi*k*n/N)) #Here is the equation implemented for DFT

plt.subplot(3,1,2)
plt.title("Fig 2: Magnitude Spectrum with Phase X(k)")
plt.plot(np.angle(X), label='Phase')
plt.plot(abs(X), label='Magnitude')
plt.legend()




#frequency Axis

fa = (Fs/2)*np.arange(0,1,2/N)    #followed nquist sampling therom
X_magnitude = (2/N) * abs(X[0:np.size(fa)])

plt.subplot(3,1,3)  
plt.title("Fig 3: Frequency Spectrum (frequency domain)")                          #frequency spectrum after dft
plt.plot(fa,X_magnitude)
plt.ylabel('Amplitude')
plt.xlabel('Frequency')

plt.subplots_adjust(hspace=.5)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 12)
```

As we can see above the code , we took three frequncy f1=40, f2=80 and f3=100. <br> By looking the frequency spectrum found by DFT, we can see that<br> We found our expected result.


# **4. Compute the inverse DFT of the signal obtained in (3).**

From 3 after doing DFT of x(n) we got X(k).<br> Now using the X(k) as input of **IDFT**, we will find out x(n).

```python
#Input of the IDFT X(k), Frequency domain signal
plt.subplot(2,1,1)
plt.title("Fig 4: Frequency Domain Signal X(k)")
plt.plot(np.angle(X), label='Phase')
plt.plot(abs(X), label='Magnitude')
plt.legend()



#IDFT
x =np.zeros(N,complex)
k=np.arange(N)
for n in range(N):
  x[n] = (1/N)*np.sum(X*np.exp(2j*np.pi*k*n/N)) #Here is the equation implemented for IDFT

plt.subplot(2,1,1)
plt.title("Fig 4: Frequency Domain Signal X(k)")
plt.plot(np.angle(X), label='Phase')
plt.plot(abs(X), label='Magnitude')
plt.legend()



plt.subplot(2,1,2)
plt.title("Fig 5: Time Domain Signal x(n)")
plt.plot(t,np.real(x))
plt.ylabel('Amplitude')
plt.xlabel('Time')


plt.subplots_adjust(hspace=.5)
plt.rc('font', size=15)
fig = plt.gcf()
fig.set_size_inches(16, 9)
```

**As we can see** ***Figure 5*** is exactly the same of ***Figure 1***.<br>
That means the sane Signal that was used as an input of DFT to get a frequency domain signal,<br>
can be exatly recovered by IDFT by taking the input of the frequency domain signal implemented here.
