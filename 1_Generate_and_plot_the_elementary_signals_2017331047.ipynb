```
import numpy as np
import matplotlib.pyplot as plt
plt.figure(figsize=(25,25))
z=np.linspace(-1.0,1.0,160)
xn=np.cos(z*10*np.pi)

lim = 1
i = 1
 
def DrawnGraph(n, x_n):
  global i
  plt.subplot(4, 1, i)
  plt.stem(n, x_n)
  plt.axvline(linewidth="1",color='r')
  plt.axhline(linewidth="1",color='r')
  i = i+1
```

# **Unit Impulse**

```
plt.figure(figsize=(25,25))
def unit_impulse(n):
  ret=np.zeros(160)
  if n>=0:
    k=80+n
  else:
    k=80-n
  for i in range(0,160):
    if i==k:
      ret[i]=1
  return ret

k=-5
imp= unit_impulse(k)
DrawnGraph(z,imp)
```

# **Unit Step**

```
plt.figure(figsize=(25,25))
def unit_step(n):
    ret=np.zeros(160)
    if(n>=0):
        k=80-n;
    else:
        k=80+n;
    for i in range(k,160):
      ret[i]=1;
    return ret

k=0
uit= unit_step(k)
DrawnGraph(z,uit)
```

# **Unit Ramp**

```
plt.figure(figsize=(25,25))
def unit_ramp():
  ret=np.zeros(160)
  kk=0
  for i in range(80,160):
     
      ret[i]=kk
      kk=kk+1
  return ret

ramp=unit_ramp()
DrawnGraph(z,ramp)
```

# **Exponential Signal**

```
plt.figure(figsize=(25,25))
import math
def exponential_signal(a,n):
  ret=np.zeros(160)
  if(n>=0):
    k=80-n;
  else:
    k=80+n;
  cnt=0
  for i in range(k,160):
    ret[i]=pow(a,cnt)
    cnt=cnt+1
  return ret
  

a=1.5
k=0
z=np.linspace(-80,80,160)
ex= exponential_signal(a,k)
DrawnGraph(z,ex)
```
